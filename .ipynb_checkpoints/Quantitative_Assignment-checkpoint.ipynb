```python
#Importing packages

!pip install yfinance
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import math
from google.colab import output   #For clearing output in case of warning given; Just to make the outputs look better
output.clear()
```

```python
#Reading Data

data=yf.download("BHARTIARTL.NS",data_source='yahoo',start='2016-08-8',end='2021-08-8')   #Data for bhartiartl(communication) is downloaded

print('-------MENU-------')
print('1 : First 3 year data')
print('2 : Last 2 year data')
print('Anything else : Full 5 year data\n')
ch=eval(input("Enter your choice="))

if ch==1:
  sample=data['Close'][:739]
elif ch==2:
  sample=data['Close'][739:]
else:
  sample=data['Close']

#Making Close Price vs Date chart
plt.figure(figsize=(15,5))
plt.title('Close Price Chart')
plt.plot(sample)    
plt.xlabel("Date")
plt.ylabel("Close Price (Rs.)")
plt.show()
```

```python
#Calculating RSI

period=14    #setting the period of calculating RSI to 14 days
delta=sample.diff()   #This will contain difference in stock closing price on a particular day and the day before, i.e. change from previous day
up=delta.clip(lower=0)    #This will contain stock price change where change was positive and for other dates, will be 0
down=-1*delta.clip(upper=0)   #This will contain stock price change where change was negative and for other dates, will be 0
ma_up=up.rolling(window=period).mean()    #Calculating the average gain
ma_down=down.rolling(window=period).mean()    #Calculating the average loss
rsi=100-(100/(1+ma_up/ma_down))   #Calculating the RSI for each day

#Making RSI vs Date Chart
plt.figure(figsize=(15,5))
plt.title('RSI chart')
plt.plot(rsi) 
plt.xlabel("Date")
plt.ylabel("RSI")
plt.axhline(70,linestyle='--',c='red',alpha=0.5)
plt.axhline(30,linestyle='--',c='green',alpha=0.5)
plt.show()
```

```python
#Finding Buy and Sell Points

dtf=sample.to_frame()
dtf['RSI']=rsi
dtf['buy_points']=np.where((rsi<30)&(rsi.shift(1)>=30), 1, 0)   #If RSI crosses 30 from above 30, we buy
dtf['sell_points']=np.where((rsi>70)&(rsi.shift(1)<=70), 1, 0)    #If RSI crosses 70 from below 70, we sell

#The number of buy and sell signals might not be equal and we might get two of the same signal one after the other
#Hence we find the position
dtf['position']=0   #The position at any date: 0 if no stocks held and 1 if stocks held
bought=0    #Variable for filling up values of position column
for i in range(sample.size):
  if dtf['buy_points'][i]==1 and bought==0:
    bought=1
  if dtf['sell_points'][i]==1 and bought==1:
    bought=0
  dtf['position'][i]=bought
output.clear()    #For clearning output which was giving a warning
```

```python
#Plotting Buy and Sell Signals on the Close Price Graph

plt.figure(figsize=(15,5))
plt.title('Buy/Sell Chart')
plt.plot(dtf['Close'])
for i in range(1,sample.size):
  if(dtf['position'][i-1]==0 and dtf['position'][i]==1):    #If position changes from 0 to 1, we have bought stocks
    plt.plot(dtf.index[i],dtf['Close'][i],'^', markersize=10, color='g')
  if(dtf['position'][i-1]==1 and dtf['position'][i]==0):    #If position changes from 1 to 0, we have sold the stocks
    plt.plot(dtf.index[i],dtf['Close'][i],'v', markersize=10, color='r')
plt.xlabel("Date")
plt.ylabel("Close Price (Rs.)")
plt.show()
```

```python
#Calculating Final Portfolio and Return %

summary_dict={'Date':[],'Signal':[],'Profit':[]}    #For making the final summary of trades with timestamp
capital=100000.0    #Initial capital is Rs. 1,00,000
buy_price,sell_price=0.0,0.0    #Will store last buy price of stocks and selling price at the end of trade
dtf['portfolio']=capital    #To keep account of our portfolio on each date
stocks=0.0    #To store number of stocks held
for i in range(1,sample.size):
  if dtf['position'][i-1]==0 and dtf['position'][i]==1:    #If position changes from 0 to 1, stocks have been bought
    buy_price=dtf['Close'][i]
    summary_dict['Date'].append(dtf.index[i])
    summary_dict['Signal'].append('Buy')
    summary_dict['Profit'].append(-capital/100)
    stocks=capital*99.0/100/dtf['Close'][i]   #Number of stocks bought
  if dtf['position'][i-1]==1 and dtf['position'][i]==0 :
    sell_price=dtf['Close'][i]
    summary_dict['Date'].append(dtf.index[i])
    summary_dict['Signal'].append('Sell')
    summary_dict['Profit'].append((capital*sell_price/buy_price*99/100-capital)*99/100)
    capital=stocks*dtf['Close'][i]*99.0/100   #Capital is calculated after selling stocks each time (it is the portfolio after last time stocks were sold)
    stocks=0
  if stocks==0:
    dtf['portfolio'][i]=capital
  else:
    dtf['portfolio'][i]=stocks*dtf['Close'][i]
  if dtf['portfolio'][i]<85000 and dtf['position'][i-10]==1:   #Setting Stop loss at 85000
    while(True):
      j=i
      if dtf['position'][j+1]==0:
        break
      dtf['position'][j+1]=0
      j=j+1 

output.clear()    #For clearing output which was giving a warning(no effect on the result)

print("\nFinal Capital= Rs.",dtf['portfolio'][-1])   #Final capital/portfolio
print('Profit= Rs.',dtf['portfolio'][-1]-dtf['portfolio'][0])
print("Return=",(dtf['portfolio'][-1]-dtf['portfolio'][0])/dtf['portfolio'][0]*100,"%")
benchmark=100000.0*sample[-1]/sample[0]    #Calculating benchmark
print("Benchmark= Rs.",benchmark)

#Plotting the Portfolio and Benchmark Chart
plt.figure(figsize=(15,5))
plt.title('Portfolio Chart')
plt.plot(dtf['portfolio'],label='portfolio')
plt.plot(dtf['Close']*100000/dtf['Close'][0],linestyle='--',alpha=0.4,label='benchmark')
plt.axhline(100000,linestyle='--',c='green',alpha=0.5,label='initial capital')
plt.xlabel("Date")
plt.ylabel("Portfolio (Rs.)")
plt.legend(loc='upper left')
plt.show()
```

```python

#Creating the required .csv file

ohclv=pd.DataFrame()
if ch==1:
  ohclv=data[:739]   #For in sample: first 3 year data
elif ch==2:
  ohclv=data[739:]    #For out sample: last 2 year data
else:
  ohclv=data    #For complete 5 year data
ohclv['RSI']=rsi
ohclv['buy_signal']=dtf['buy_points']
ohclv['sell_signal']=dtf['sell_points']
ohclv['position']=np.where(dtf['position']==1,'long','neutral')
ohclv['portfolio']=dtf['portfolio']
ohclv.to_csv("Summary.csv")
output.clear()
```

```python
#Finding Data required for Summary.doc file

print("Total number of buy signals generated=",dtf['buy_points'][dtf['buy_points']>0].size)
print("Total number of sell signals generated=",dtf['sell_points'][dtf['sell_points']>0].size)
summary=pd.DataFrame(summary_dict)    #Creating Summary DataFrame from dictionary created earlier
print("Total number of trades executed=",summary.index.size)
print("Profit for period=Rs.",dtf['portfolio'][-1]-100000)
print('Annualized Return=',(math.pow((1+(dtf['portfolio'][-1]-100000)/100000),(365.0/sample.index.size))-1)*100,'%')

#Calculating win/loss percentage
win,loss=0,0
for i in summary['Profit']:
  if i>0:   #if profit>0 then win else loss
    win=win+1
  elif i<0:
    loss=loss+1
loss=loss-math.floor(summary['Date'].size/2)    #Not considering transaction loss during each buy signal
print("Win/Loss % for the period=",win/loss*100,'%')

#Calculating maximum drawdown
mdd=0.0
max=0.0
for i in dtf['portfolio']:
  if max<i:
    max=i
  if mdd<(max-i)/max:
    mdd=(max-i)/max
print("Maximum Drawdown=",mdd*100,'%')

print("\nSUMMARY:")
summary
```
